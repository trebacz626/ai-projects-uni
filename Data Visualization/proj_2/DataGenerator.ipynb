```python
from plotnine import *
import pandas as pd
import numpy as np
```

```python
biology = pd.read_csv("Sugars.csv")
biology.head()
```

```python
array = np.array(biology.columns)
angles = list(biology.columns)
angles = [x for x in list_of_bond_lengths if x.count("-") == 2 and x.count("/")==0]
angles
```

```python
df = pd.DataFrame(columns=["AngleName","Sugar", "Value"])
for angle in angles:
    to_append = biology[["Sugar",angle]].set_index("Sugar").stack().reset_index().rename(columns={'level_1':'AngleName', 0:'Value'}).dropna(subset=["Value"])
    df = pd.concat([df,to_append])
#df = df.set_index("AngleName")
df.describe()
df.to_csv("Angle.csv", index=False)
```
